```python
import pandas as pd
from matplotlib import pyplot as plt
# creating DataFrame from csv file, droping 1st row, then set 'Data Sourse' row as a indexes, then replace all NaN in our DataFrame by 0
df = pd.read_csv('/home/hp/Downloads/data_sets/countries_debt/API_DT.AMT.DLXF.CD_DS2_en_csv_v2.csv').drop([0]).set_index('Data Source').fillna(0)

# setting 1st row as a head of our DataFrame
new_columns_names = list(df.iloc[0].values)
df.set_axis(new_columns_names, axis = 'columns', inplace=True)

# drop a row
df.drop(['Country Name'], inplace = True)

# droping some coloumns 
df.drop(['Country Code', 'Indicator Name', 'Indicator Code',1960.0,1961.0,1962.0,1963.0,1964.0,1965.0,1966.0,1967.0,1968.0,1969.0], axis='columns' ,inplace = True)

# print all countries names in our Data in acciding order
for i in sorted([i for i in df.index]): print(i, end='//')
```

```python
def international_Debt_History(country_name, x_size=12, y_size=6):
    '''
    return a graph of international Debt History for spacific country 
    '''
    # selecting a spacific country and drop first 7 columns data
    country = df.loc[country_name]    
    debt = list(country.values)
    years = list(country.keys())
    plt.figure(figsize=(x_size,y_size))
    plt.bar(years, debt)
    plt.title('{} international Debt History'.format(country_name), size = 25, color= 'r')
    plt.ylabel('1000 Million $', size = 25, color='r')
    plt.xticks(years, rotation=90, color='r')
    plt.grid(True)
    plt.text(1970, max(debt), 'Data Sourse: World bank website [https://data.worldbank.org/indicator/DT.AMT.DLXF.CD]', color='b', size = 12)
    plt.savefig('{}.svg'.format(country_name))
    plt.show()
lst = ['Pakistan', 'India', 'Brazil', 'Turkey', 'United States']
for i in lst:
    international_Debt_History(i)
```
