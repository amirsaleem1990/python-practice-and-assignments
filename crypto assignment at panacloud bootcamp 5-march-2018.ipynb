# Crypto assingment

```python
import requests as r
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
def crypto_defrence(fiat = 'btc', crypto = 'usd'):
    url = 'https://api.cryptonator.com/api/full/{}-{}'.format(fiat, crypto)
    df = pd.DataFrame(r.get(url).json()['ticker']['markets'])
    df['price']  = df['price'].astype('float')
    y, x , n=list(df['volume']),list(df['price']),list(df['market'])  
    plt.figure(figsize=(20,10)), plt.scatter(x, y, s=150)
    for i, txt in enumerate(n): plt.annotate(txt, (x[i],y[i])) 
    plt.rc('xtick', labelsize=20) 
    plt.rc('ytick', labelsize=20) 
    plt.rcParams.update({'font.size': 22})
    print('minimum price at:\n',end='');print((df[df['price'] == np.min(df['price'])]))
    print('\nmaximum price at:\n');print((df[df['price'] == np.max(df['price'])]))
    print('\ndefrence between minimum and meximum:\n',end=''); print(np.max(df['price']) - np.min(df['price']))
    plt.show(); return ' '
print(crypto_defrence('btc', 'usd'))
```
