```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
df = pd.read_csv('train.csv')
df.head()
```

```python
male = len(df.query('Sex == "male"'))
female = len(df.query('Sex == "female"'))
total = [male, female]
plt.figure(figsize=(3,6))
bar = plt.bar(['Male\'s', 'female\'s'], [male, female], color=['b', 'pink'])
plt.xticks(size= 20)
plt.ylabel('Qty', size= 30, color = 'brown')
plt.title('Males and Females QTY. and %', size=30, color='brown')
for i in bar: plt.text(i.get_x(), i.get_height(),str(round((total[bar.index(i)]/sum(total)*100),2))+'%', size=20,)
```

```python
av_age = np.mean(df.Age)
print('avg. age for all passegers :','-'*2, round(av_age, 2))
av_males_age = np.mean(df.query('Sex == "male"').Age)
print('avg. age for Males :','-'*10, round(av_males_age,2))
av_females_age = np.mean(df.query('Sex == "female"').Age)
print('avg. age for Females: ', '-'*8, round(av_females_age,2))
```

```python
av = plt.bar(['avg. age', 'avg. male age', 'avg. female age'], [av_age, av_males_age, av_females_age])
plt.xticks(rotation = 45, size=20)
for i in av: 
    plt.text(i.get_x(),i.get_height()-5,str(round(i.get_height(),2)), size=20)
plt.ylabel('age', color='r', size = 40)
plt.title('average ages', size = 20, color='r')
```

```python
c1 = len(df.query('Pclass == 1'))
c2 = len(df.query('Pclass == 2'))
c3 = len(df.query('Pclass == 3'))
clas = plt.bar(['class1', 'class2', 'class3'], [c1,c2,c3])
plt.xticks(size=20)
plt.ylabel('passengers Qty.', color='r', size = 20)
plt.title('Distibution by Class', color = 'r', size = 20)
heigh = [i.get_height() for i in clas]
c11 = len(df.query('Pclass == 1 & Survived == 1'))
c22 = len(df.query('Pclass == 2 & Survived == 1'))
c33 = len(df.query('Pclass == 3 & Survived == 1'))
clas2 = plt.bar(['class1', 'class2', 'class3'], [c11,c22,c33], width= 0.6)
plt.xticks(size=20)
plt.ylabel('passengers Qty.', color='r', size = 20)
plt.title('Distibution by Class', color = 'r', size = 20)
plt.legend(['Total passengers', 'Survived only']) #
for i in clas2: plt.text(i.get_x(), i.get_height(), str(round(i.get_height()/heigh[clas2.index(i)]*100,2))+'%')
```

```python
a = ['male\'s in class1', 'male\'s in class2', 'male\'s in class3', 'female\'s in class1', 'female\'s in class2', 'female\'s in class3']
male_c1 = len(df.query('Sex == "male" & Pclass == 1')) 
male_c2 = len(df.query('Sex == "male" & Pclass == 2'))
male_c3 = len(df.query('Sex == "male" & Pclass == 3'))
female_c1 = len(df.query('Sex == "female" & Pclass == 1'))
female_c2 = len(df.query('Sex == "female" & Pclass == 2'))
female_c3 = len(df.query('Sex == "female" & Pclass == 3'))
plt.figure(figsize=(4, 6))
aa = plt.bar(a, [male_c1, male_c2, male_c3, female_c1, female_c2, female_c3],width = 0.8)
plt.xticks(rotation = 90, size = 15)
aa_height = []
for i in aa: 
    plt.text(i.get_x(), i.get_height(),str(i.get_height()));
    aa_height.append(i.get_height())
S_m_c1 = len(df.query('Sex == "male" & Pclass == 1 & Survived == 1'))
S_m_c2 = len(df.query('Sex == "male" & Pclass == 2 & Survived == 1'))
S_m_c3 = len(df.query('Sex == "male" & Pclass == 3 & Survived == 1'))
S_fm_c1 = len(df.query('Sex == "female" & Pclass == 1 & Survived == 1'))
S_fm_c2 = len(df.query('Sex == "female" & Pclass == 2 & Survived == 1'))
S_fm_c3 = len(df.query('Sex == "female" & Pclass == 3 & Survived == 1'))
b = plt.bar(a, [S_m_c1, S_m_c2, S_m_c3, S_fm_c1, S_fm_c2, S_fm_c3], width = 0.6)
plt.ylabel('Passengers Qty.', size=20, color='r')
plt.legend(['Total passengers', 'Survived only'])
for i in b: 
    plt.text(i.get_x(), i.get_height()-15,str(i.get_height()))
    plt.text(i.get_x(), i.get_height()+15,str(round((aa_height[b.index(i)])-i.get_height())), color='r')
```
